```python
import pandas as pd
df=pd.read_csv('/content/mtcars.csv')
print(df.head ())
print(df.tail ())
print(df.info ())
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
auto_count = (df['am'] == 0).sum()
manual_count = (df['am'] == 1).sum()
labels = ['Automatic', 'Manual']
counts = [auto_count, manual_count]
plt.bar(labels, counts, color=['green', 'pink'])
plt.title("Frequency of Transmission Types")
plt.xlabel("Transmission Type")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.scatter(df['wt'], df['mpg'], color='green')
plt.title("Car Weight vs MPG")
plt.xlabel("Weight (1000 lbs)")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```

```python
auto = (df['am'] == 0).sum()
manual = (df['am'] == 1).sum()
plt.bar(['Automatic', 'Manual'], [auto, manual], color=['pink','red'])
plt.title("Frequency of Transmission Types")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.boxplot(df['mpg'])
plt.title("Box Plot of MPG")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```
